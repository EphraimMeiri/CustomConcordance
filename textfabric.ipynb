```python
# join the lexemes by root
lexeme_by_root = {}
for lex in all_lexemes:
    root = lex[1]
    if root not in lexeme_by_root:
        lexeme_by_root[root] = []
    lexeme_by_root[root].append(lex)
lexeme_by_root
```

```python
def heb_num(num):
    heb = ['×','×‘','×’','×“','×”','×•','×–','×—','×˜','×™','×™×','×™×‘','×™×’','×™×“','×˜×•','×˜×–','×™×–','×™×—','×™×˜','×›','×›×','×›×‘','×›×’','×›×“','×›×”','×›×•','×›×–','×›×—','×›×˜','×œ','×œ×']
    return heb[num-1]
def tuple_to_str(t):
    return f"{heb_num(t[0])}:{heb_num(t[1])}"
```

```python
# Output to string
output = ""
roots = sorted(lexeme_by_root.keys())
for root in roots:
    output+= root+'\n'
    for lex in lexeme_by_root[root]:
        output+= "\t"+lex[0]+" "+tuple_to_str(lex[3])+ "\n"
    output+= "\n"
```
